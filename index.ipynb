## Employees and their Office (a One-to-One join)

Return a list of all of the employees with their first name, last name and the city and state of the office that they work out of (if they have one). Include all employees and order them by their first name, then their last name.

```python
import pandas as pd
import sqlite3
con = sqlite3.connect('data.sqlite', detect_types = sqlite3.PARSE_COLNAMES)
cur = con.cursor()
#Your code here
```

```python
cur.execute('''SELECT * FROM offices
                JOIN employees USING (officeCode);
                ''')
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df = df[['firstName', 'lastName', 'city', 'state']]
df.sort_values(by = ['firstName'])
```

## Customers and their Orders (a One-to-Many join)

Return a list of all of the customers. For each customer return a record for each of their order numbers, order dates and statuses.

```python
cur.execute('''SELECT * FROM customers
                JOIN orders USING (customerNumber);
                ''')
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df = df[['customerName', 'orderNumber', 'orderDate', 'status']]
df.head()

def info(dataframe):
    sales_dict = {}
    for index, row in dataframe.iterrows():
        if row['customerName'] not in sales_dict:
            sales_dict[row['customerName']] = [row['orderDate'], row['orderNumber'], row['status']]
        else:
            sales_dict[row['customerName']].append([row['orderDate'], row['orderNumber'], row['status']])
            
    return sales_dict

info(df)
```

## Orders and their Order Details (another One-to-Many join)

Return a list of orders. For each order return a record for each order detail within the order.

```python
cur.execute('''SELECT * FROM orders
                JOIN orderdetails USING (orderNumber);
                ''')
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]

def info(dataframe):
    sales_dict = {}
    for index, row in dataframe.iterrows():
        if row['orderNumber'] not in sales_dict:
            sales_dict[row['orderNumber']] = [row['productCode'], row['quantityOrdered'], row['priceEach'], row['orderLineNumber']]
        else:
            sales_dict[row['orderNumber']].append([row['productCode'], row['quantityOrdered'], row['priceEach'], row['orderLineNumber']])
            
    return sales_dict

info(df)
```

## Orders, Order details and Product Details (a Many-to-Many Join)

Return a list of the orders with the the order information, and allthe product information for each product in the order. 

```python
cur.execute('''SELECT * FROM orders
                JOIN orderdetails USING (orderNumber)
                JOIN products USING (productCode);
                ''')
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
print(len(df))
df.head()
```
